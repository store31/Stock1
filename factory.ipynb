```python

import os
import asyncio

video_list = [
    {"script": "AI is transforming the tech world.", "prompt": "Futuristic robot working on a laptop, 4k"},
    {"script": "The beauty of deep space is endless.", "prompt": "Cinematic nebula in deep space, stars and planets"},
    {"script": "Nature is the best artist.", "prompt": "Hyper-realistic waterfall in a green forest"}
]

async def make_us_video(script, prompt):
    print("🎥 (تجربة) صنع فيديو:", script)
    await asyncio.sleep(1)
    with open("video_final.mp4", "w") as f:
        f.write("fake video")

async def main():
    print(f"🚀 بدأت الماكينة في صنع {len(video_list)} فيديوهات...")

    for i, item in enumerate(video_list):
        print(f"\n🎬 جاري العمل على الفيديو {i+1}...")

        await make_us_video(item['script'], item['prompt'])

        os.rename("video_final.mp4", f"short_video_{i+1}.mp4")

        print(f"✅ تم حفظ الفيديو: short_video_{i+1}.mp4")

    print("\n🎉 مبروك! قاع الفيديوهات واجدين.")

await main()
```
